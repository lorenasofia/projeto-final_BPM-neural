```python
traducoes = {
    'title': 'musica',
    'artist': 'artista',
    'top genre': 'genero',
    'year': 'ano',
    'bpm': 'beat',
    'energy': 'sentimento',
    'danceability': 'dancante',
    'db': 'volume',
    'liveness': 'vivacidade',
    'valence': 'afetividade',
    'duration' : 'duracao',
    'acousticness': 'acustica',
    'speechiness' : 'expressividadeVocal',
    'popularity': 'popularidade'
}

novas_colunas = []

for chave, valor in traducoes.items():
    novas_colunas.append(valor)

novas_colunas
```

- **dB (Decibéis):** Isso geralmente representa o volume ou nível de som de uma música. Valores negativos de dB indicam volume mais baixo, enquanto valores positivos indicam volume mais alto.

- **liveness (Vivacidade):** Refere-se à probabilidade de uma música ter sido gravada ao vivo em vez de em um estúdio. Pode ser uma métrica binária (ao vivo/não ao vivo) ou uma pontuação que indica o quão "ao vivo" a música soa.

- **valence:** Refere-se à positividade ou negatividade da música. Músicas com alto valor de valence são mais alegres e positivas, enquanto músicas com valor baixo são mais tristes ou negativas.

- **acousticness (Acústica):** Indica o quão "acústica" uma música é. Valores mais altos indicam uma música mais acústica, enquanto valores mais baixos indicam uma música mais eletrônica ou processada.

- **speechiness (Expressividade vocal):** Refere-se ao grau de presença de palavras faladas ou vocalizações na música. Valores mais altos indicam uma música com mais palavras faladas.

- **popularity (Popularidade):** Representa o quão popular ou bem-sucedida é uma música, frequentemente com base em classificações, reproduções ou outras métricas.

```python
df.columns = novas_colunas
df
```
